```python
import pandas as pd

DATA_DIR = '/content/drive/MyDrive/RL_recommendation-zaid/DATA/Research dataset_2.xlsx'
df =  pd.read_excel(DATA_DIR)
df.head()
```

```python
#drop username and Unnamed: 14 columns
df = df.drop(['username', 'Unnamed: 14'], axis=1)
df.head()
```

```python
#rename column
df = df.rename(columns={'Possible output to overcome the negative emotion currently experienced': 'suggestion'})
df.head()
```

```python
import json

json_records = []
for index, row in df.iterrows():
    json_records.append(row.to_dict())
```

```python
type(json_records[0])
```

```python
def generate_query_from_json_excluding(user_details, suggestion):
    query = "Suppose there is a user with the following details:"
    for key, value in user_details.items():
        if key != suggestion:
            query += f'"{key}": "{value}",'
    query += f'A System suggesting him to {suggestion}.'
    query += "Generate a short paragraph for him about this suggestion."

    return query
```

```python
generate_query_from_json_excluding(json_records[0], json_records[0]['suggestion'])
```

```python
from openai import OpenAI
client = OpenAI(
    api_key = ''
)

def chat_with_gpt(user_input= "hi Good Evening"):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": user_input}]
    )
    content = response.choices[0].message.content
    return str(content)
```

```python
prompt = generate_query_from_json_excluding(json_records[0], json_records[0]['suggestion'])
chat_with_gpt(prompt)
```

```python
import time

rows = []
for json_data in json_records[:30]:
    row_data = {}
    prompt = generate_query_from_json_excluding(json_data, json_data['suggestion'])
    row_data['prompt'] = prompt
    response = chat_with_gpt(prompt)
    row_data['text'] = response

    rows.append(row_data)
    time.sleep(5)
new_df = pd.DataFrame(rows)
```
